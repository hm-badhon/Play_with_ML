# Feature Importance techniques

```python
x
```

```python
x2 = x.drop(['cardio'],axis=1)
```

```python
x2
```

```python
y
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
```

```python
model.fit(x2,y)
```

```python
model.feature_importances_
```

```python
# Series function

top= pd.Series(model.feature_importances_,index = x2.columns)
```

```python
top
```

```python
top.nlargest(8).plot(kind = 'bar')
```

```python
top.nlargest(8).plot(kind = 'barh')
```
